```python
import pandas as pd
import sqlalchemy as db
```

```python
conn = pd.read_json('connection.json', typ='series')
host = conn['host']
user = conn['user']
password = conn['password']
port = conn['port']

db_name = input('DB Name: ')
engine = db.create_engine('mysql+pymysql://' + user + ':' + password + '@' + host + ':' + port + '/' + db_name)
```

```python
data = pd.read_sql('SELECT * FROM `rp9921-2022`', engine)
```

```python
data
```

```python
df = pd.DataFrame()
offsetError = 0
i = 0
year = 2022

def query(hourNum, year, data):
    global offsetError
    global df
#     query = f"SELECT `index`, ID, Time, Date, FF_1, FF_2 FROM `{acft_model}-{year}` LIMIT 3600 OFFSET {hourNum * 3600 - offsetError};"
    index = hourNum * 3600 - offsetError
    new_df = data[index : index + 3600]
    new_df.reset_index(inplace=True)
    if new_df['ID'].nunique() > 1:
        id_select = min(new_df['ID'].unique())
        f_df = new_df[new_df['ID'] == id_select]
        offsetError += len(new_df[new_df['ID'] == max(new_df['ID'].unique())])
        new_df = f_df
    ID = min(new_df['ID'].unique())
    ff1 = sum(new_df['FF_1']) / 3600
    ff2 = sum(new_df['FF_2']) / 3600
    Date = max(pd.to_datetime(new_df['Date']))
    year = Date.year
    month = Date.month
    date = Date.day
    min_hour = min(pd.to_datetime(new_df['Time']).dt.hour)
    max_hour = max(pd.to_datetime(new_df['Time']).dt.hour)
    length = len(new_df)
    min_time = new_df['Time'][0]
    max_time = new_df['Time'][length-1]
    if min_hour == 0 and max_hour == 23:
        max_hour = 0
    create_df = {'ID':ID, 'year':year, 'month':month, 'date':date, 'hour':max_hour, 'ff1':ff1, 'ff2':ff2, 'min_time': min_time, 'max_time': max_time, 'len':length}
    create_df = pd.DataFrame(create_df, index=[0])
    df = pd.concat([df, create_df])
    print(hourNum, ID, year, month, date, max_hour, min_time, max_time, length, offsetError)

print("hourNum, ID, year, month, date, max_hour, min_time, max_time, length, offsetError")
while True:
    try:
        query(i, year, data)
        i += 1
    except:
        print("Hournum: ", i)
        print("Offset: ", offsetError)
        break
```

```python
def complete_hr(last_hr):
    if last_hr == 23:
        return 0, True
    else: 
        return last_hr + 1, False

def complete_date(last_date, last_month, last_year):
    # feb 28 for both 2022-2023    
    isAddMonth = False
    isAddYear = False
    days_month = [31,28,31,30,31,30,31,31,30,31,30,31]
    max_day = days_month[last_month-1]
    new_date = last_date
    new_month = last_month
    new_year = last_year
    
    if last_date == max_day:
        new_date = 1
        isAddMonth = True
    else:
        new_date = last_date + 1
    
    if isAddMonth:
        if last_month == 12:
            new_month = 1
            isAddYear = True
        else:
            new_month = last_month + 1
    
    if isAddYear:
        new_year = last_year + 1
        
    return new_date, new_month, new_year

for i in range(len(hist)):
    if i > 0:
        prev_ID = hist.iloc[i-1,0]
        prev_hour = hist.iloc[i-1,4]
        curr_ID = hist.iloc[i,0]
        curr_hour = hist.iloc[i,4]
        curr_date = hist.iloc[i,3]
        curr_month = hist.iloc[i,2]
        curr_year = hist.iloc[i,1]
        if prev_ID == curr_ID and prev_hour == curr_hour:
            new_hour, isAddDate = complete_hr(curr_hour)
            hist.iloc[i,4] = new_hour
            if isAddDate:
                new_date, new_month, new_year = complete_date(curr_date, curr_month, curr_year)
                hist.iloc[i, 3] = new_date
                hist.iloc[i, 2] = new_month
                hist.iloc[i, 1] = new_year
```

```python
min_year, max_year = min(hist['year']), max(hist['year'])
min_month, max_month = min(hist[hist['year'] == min_year]['month']), max(hist[hist['year'] == max_year]['month'])
min_date, max_date = min(hist[hist['month'] == min_month]['date']), max(hist[hist['month'] == max_month]['date'])
min_hour, max_hour = min(hist[(hist['year'] == min_year) & (hist['month'] == min_month)  & (hist['date'] == min_date)]['hour']), max(hist[(hist['year'] == max_year) & (hist['month'] == max_month) & (hist['date'] == max_date)]['hour'])
min_datetime = f"{min_month}-{min_date}-{min_year} {min_hour}:00:00"
max_datetime = f"{max_month}-{max_date}-{max_year} {max_hour}:00:00"
print("Min Date:", min_datetime)
print("Max Date:", max_datetime)
date_range = pd.date_range(min_datetime, max_datetime, freq='H')
```

```python
new_hist = pd.DataFrame({'ff': 0}, index=date_range).reset_index()
for each in hist.iloc:
    year = int(each['year'])
    month = int(each['month'])
    date = int(each['date'])
    hour = int(each['hour'])
    datetime = f"{year}-{month}-{date} {hour}:00:00"
    index = pd.to_datetime(datetime)
    new_index = new_hist[new_hist['index'] == index].index.to_list()[0]
    new_hist.iloc[new_index, 1] = each['ff']
```

```python
new_hist.index = new_hist['index']
if 'index' in new_hist.columns:
    new_hist.drop('index', axis=1, inplace=True)
```

```python
months = sorted(list(set([date.strftime('%Y-%m') for date in new_hist.index])))
ticks = []
labels = []
for i in range(len(new_hist)):
    date = (str(new_hist.index[i])[:7])
    if date in months:
        ticks.append(pd.to_datetime(date))
        labels.append(date)
        months.remove(date)
```

```python
ax = new_hist.ff.plot(figsize=(50,10), xticks=ticks)
ax.set_xticklabels(labels)
```
